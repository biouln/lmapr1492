```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
import matplotlib.pyplot as plt
```

```python
with MPRester("tA2HQFkWnG7wRPHf") as mp:
    # Create a structure for material id mp-23156 (He)
    structure = mp.get_structure_by_material_id("mp-23156")
    print(structure)
```

```python
xrdcal = XRDCalculator('CuKa')
pattern = xrdcal.get_pattern(structure)
```

```python
print(pattern.hkls)
```

```python
print("First Spike of the difractogram for a CuKa incident ray of wave length λ=1.54060 Å")
print("-------------------------------------------------------------------------------------------------")
print(" ")
print("Angle 2\u0398 [°] :", pattern.x[0])
print("Intensity [%] :", pattern.y[0])
print("Associated Miller Indices :", pattern.hkls[0])
print("                   d :", pattern.d_hkls[0])
```

```python
print("Second Spike of the difractogram for a CuKa incident ray of wave length λ=1.54060 Å")
print("-------------------------------------------------------------------------------------------------")
print(" ")
print("Angle 2\u0398 [°] :", pattern.x[1])
print("Intensity [%] :", pattern.y[1])
print("Associated Miller Indices :", pattern.hkls[1])
print("                   d :", pattern.d_hkls[1])
```

```python
print("Third Spike of the difractogram for a CuKa incident ray of wave length λ=1.54060 Å")
print("-------------------------------------------------------------------------------------------------")
print(" ")
print("Angle 2\u0398 [°] :", pattern.x[2])
print("Intensity [%] :", pattern.y[2])
print("Associated Miller Indices :", pattern.hkls[2])
print("                   d :", pattern.d_hkls[2])
```

```python
print("Four first difractogram spikes for a CuKa incident ray of wave length λ=1.54060 Å")
xrdcal.show_plot(structure,two_theta_range=(38,45),annotate_peaks=True,ax=None,with_labels=True)
```
