```python
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
from pymatgen.symmetry.analyzer import *
from pymatgen.io.cif import *
from pymatgen.core.operations import *
from pymatgen.symmetry.site_symmetries import *
import matplotlib.pyplot as plt
import matplotlib.image as img
```

```python
with MPRester("tA2HQFkWnG7wRPHf") as mp:
    # Create a structure for material id mp-23156 (He)
    structure = mp.get_structure_by_material_id("mp-23156")
    print(structure)
    
```

```python
# Symmetry operations
a = SpacegroupAnalyzer(structure).get_symmetry_operations()
Sym_1 = a[3]
Sym_2 = a[6]
Sym_3 = a[9]

# Atomes chosen for symmetry operations (Only two available)
He_1 = [0.33333,0.66667,0.25000]#1
He_2 = [0.66667,0.33333,0.75000]#2
He_3 = [0.33333,0.66667,0.25000]#1

# Application des opérations de symétrie
He_1op = Sym_1.operate(He_1)
He_2op = Sym_2.operate(He_2)
He_3op = Sym_3.operate(He_3)
```

```python
print("First operation")
print("------------------------------")
print(" ")
print("First atome : He", He_1)
print(" ")
print("Symmetry operation number 4 : 6_bar axis : ")
print(Sym_1)
print(" ")
print("Atome after operation : He", He_1op)
print(" ")
print("Image :")

image_1=img.imread('He_Op1.jpg')
fig, axes = plt.subplots(figsize=(10,10))
axes.get_xaxis().set_visible(False)
axes.get_yaxis().set_visible(False)
axes.imshow(image_1)
plt.show()
```

```python
print("Second operation")
print("------------------------------")
print(" ")
print("First atome : He", He_2)
print(" ")
print("Symmetry operation number 7 : 2_fold screw axis : ")
print(Sym_2)
print(" ")
print("Atome after operation : He", He_2op)
print(" ")
print("Image :")

image_1=img.imread('He_Op2.jpg')
fig, axes = plt.subplots(figsize=(10,10))
axes.get_xaxis().set_visible(False)
axes.get_yaxis().set_visible(False)
axes.imshow(image_1)
plt.show()
```
